#### we can drop the column aerosols due to extreme outliers or
we can remove the outliers by doing outlier treatment


```python
df.describe()
```

```python
def capping(df):
    #calculate q1,q3
    q5=np.percentile(df,5)
    q25=np.percentile(df,25)
    q75=np.percentile(df,75)
    q95=np.percentile(df,95)
    #calculate iqr ,ll ,ul
    iqr=q75-q25
    ll=q25-1.5*iqr
    ul=q75+1.5*iqr
    #apply using lambda funct
    return df.apply(lambda y:q95 if y>ul else y).apply(lambda y:q5 if y<ll else y)
```

```python
for i in df:
    df[i]=capping(df[i])
```

```python
plt.figure(figsize=(20,10))
x=1
for i in df.columns:
    plt.subplot(4,3,x)
    sns.boxplot(df[i])
    x+=1
plt.tight_layout()
```

### lets separate out target variable from independant features

```python
x=df.iloc[:,:-2]
y=df['Temp']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
x_train
```

## As we are doing regression problem we will do feature scaling 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_trains=scaler.fit_transform(x_train)
```

```python
x_tests=scaler.fit_transform(x_test)
```

## Model building 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
lnr=LinearRegression()
lnr.fit(x_trains,y_train)
```

```python
mse=cross_val_score(lnr,x_trains,y_train,scoring='neg_mean_squared_error',cv=10)
```

```python
np.mean(mse)
```

```python
# prediction
lnr_pred=lnr.predict(x_tests)
```

```python
lnr_pred
```

```python
sns.displot(lnr_pred-y_test,kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(lnr_pred,y_test)
score
```

## Decision tree regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr=DecisionTreeRegressor(splitter='best',max_depth=4)
dtr.fit(x_trains,y_train)
```

```python
dtr_y_pred=dtr.predict(x_tests)
```

```python
sns.displot(dtr_y_pred-y_test,kind='kde')
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
dt_mse=mean_squared_error(dtr_y_pred,y_test)
dt_mae=mean_absolute_error(dtr_y_pred,y_test)
dt_r2=r2_score(dtr_y_pred,y_test)
```

```python
print(f'mean_squared_error for dt:-{dt_mse}')
print(f'mean_absolute_error for dt:-{dt_mae}')
print(f'r2_score for dt:-{dt_r2}')
```
